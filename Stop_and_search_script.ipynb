```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
import datetime as dt
import numpy as np

force = "cleveland"
response = requests.get(f"https://data.police.uk/api/stops-force?force={force}&date=2021-03")
df1 = pd.json_normalize(response.json())
```

```python
df1["age_range"].value_counts().plot(kind="bar")
plt.title("Total number of arrests by age range")
plt.ylabel("Total number of arrests in March 2021")
plt.show()
```

```python
response = requests.get(f"https://data.police.uk/api/stops-force?force={force}&date=2020-03")
df2 = pd.json_normalize(response.json())

fig, ax = plt.subplots()
ax.bar(['March 2021', 'March 2020'], [df1.shape[0], df2.shape[0]])

plt.title("Comparing the difference in total number of stop and search in March 2021 and 2020")
plt.ylabel("Total number of arrests")
plt.show()
```

```python
plt.style.use("bmh")
df1["gender"].value_counts().plot(kind="bar", title = "Total number of arrested people by gender distribution in March 2021")
plt.show()

df1["gender"].value_counts().plot(kind="bar", title = "Total number of arrested people by gender distribution in March 2020")
plt.show()
```

```python
response = requests.get(f"https://data.police.uk/api/stops-force?force={force}&date=2020-04")
df3 = pd.json_normalize(response.json())
response = requests.get(f"https://data.police.uk/api/stops-force?force={force}&date=2020-05")
df4 = pd.json_normalize(response.json())


df0 = df2[(df2["age_range"]=="over 34") & (df2["outcome"]=="Arrest")]
df00 = df3[(df3["age_range"]=="over 34") & (df3["outcome"]=="Arrest")]
df000 = df4[(df4["age_range"]=="over 34") & (df4["outcome"]=="Arrest")]

fig, ax = plt.subplots()
ax.bar(['March 2020', 'April 2020', "May 2020"], [df0.shape[0], df00.shape[0], df000.shape[0]])

plt.title("Number of adults (over 34) arrested in 2020")
plt.ylabel("Total number of arrests")
plt.show()
```
